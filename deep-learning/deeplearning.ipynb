```python
import torch
import numpy as np
import math
import time
```

```python
class Timer:
    def __init__(self):
        self.times=[]
        self.start()
    def start(self):
        self.tik=time.time()
    def stop(self):
        self.times.append(time.time()-self.tik)
        return self.times[-1]
    def avg(self):
        return sum(self.times)/len(self.times)
    def sum(self):
        return sum(self.times)
    def cumsum(self):
        return np.array(self.times).cumsum().tolist()
        
```

```python
n=10000
a=torch.ones([n])
b=torch.ones([n])
c=torch.zeros(n)
timer=Timer()
for i in range(n):
    c[i]=a[i]+b[i]
f'{timer.stop():.5f} sec'
```
