```python
import pandas as pd
```

```python
df=pd.read_csv("spotify-top-200-dataset.csv",sep=';')
df=df[df['pivot']==0]
df=df[['week','track_id','track_name','artist_names','rank','streams']]
df=df.rename(columns={
    'week':'date',
    'artist_names':'artist',
    'rank':'position'    
})
df['date']=pd.to_datetime(df['date'],format='%m/%d/%Y')
df=df.sort_values(['date','position'])
df=df[df['position']<16].reset_index(drop=True)
df['week']=df['date'].dt.isocalendar().week
df['name_short'] = (df['track_name']
    .str.replace(r'\s*[-–]\s.*$', '', regex=True)                            # 砍 ' - ' 及之后(Remix/Live/From"…")
    .str.replace(r'\s*\((feat\.|with )[^)]*\)', '', regex=True, case=False)  # 砍 (feat…)/(with…)
    .str.strip())

df[df['artist'].str.contains('Dua Lipa', case=True, na=False)].sort_values(['position','date'])
```
