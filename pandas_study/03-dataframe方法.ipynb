```python
df.groupby('Sex').agg(
    生还率   = ('Survived', 'mean'),
    平均年龄  = ('Age', 'mean'),
    最高票价  = ('Fare', 'max'),
    人数     = ('PassengerId', 'count')
)
```

```python
df['Age']=df['Age'].fillna(
    df.groupby('Pclass')['Age'].transform('mean')
)
```

```python
df['Fare_zscore'] = df.groupby('Pclass')['Fare'].transform(
    lambda x: (x - x.mean()) / x.std()
)
df['Fare_zscore']
```

```python
df.groupby('Embarked').filter(lambda x: len(x) > 100)#返回原始行
```

```python
df.groupby('Pclass').filter(
    lambda x: x['Survived'].mean() > 0.4
)#剔除整个第三仓
```

```python
df.groupby('Pclass').apply(
    lambda x: x.nlargest(3, 'Fare')
)[[ 'Name', 'Pclass', 'Fare']]
```

```python
def survival_summary(group):
    return pd.Series({
        '人数':    len(group),
        '生还人数': group['Survived'].sum(),
        '生还率':  group['Survived'].mean().round(3),
        '平均年龄': group['Age'].mean().round(1)
    })

df.groupby('Sex').apply(survival_summary)
```
