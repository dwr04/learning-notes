```python
import pandas as pd
url = "https://raw.githubusercontent.com/utdata/rwd-billboard-data/main/data-out/hot-100-current.csv"
df = pd.read_csv(url)
df=df[['chart_week','current_week','title','performer']]#去掉无用列
df=df[df['current_week']<16]#只留TOP15
df['chart_week']=pd.to_datetime(df['chart_week'])#日期改变类型
df=df[df['chart_week'].dt.year>2009]
df['week_of_year'] = df['chart_week'].dt.isocalendar().week#新增当前周
df = df.sort_values(['chart_week','current_week'], ascending=True)#排序
df=df.reset_index(drop=True)
df_backup = df.copy()# 把清洗好的最终版 df 复制一份，封存为备份

```
