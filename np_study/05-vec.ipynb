## 线性代数

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import math
from IPython.display import Image, display
```

```python
u=np.array([5,1,3])
m1=np.array([4,5,1])
m2=np.array([5,1,5])
#计算夹角
def cosine_similarity(a,b):
    a = np.array(a)
    b = np.array(b)
    denominator = np.linalg.norm(a) * np.linalg.norm(b)
    if denominator==0:
        print("error")
    else:
        return np.dot(a,b)/denominator

print(cosine_similarity(u,m1))
print(cosine_similarity(u,m2))
```

```python
print(np.cross(u, m1))  # [-14   7  21]
print(np.cross(m1, u))  # [ 14  -7 -21]
```

```python
A=np.array([[1,2,1],[3,7,2],[2,2,1]])
b=np.array([8,23,9]).reshape(-1,1)
print(np.linalg.matrix_rank(A))
print(np.linalg.matrix_rank(np.hstack([A,b])))
print(np.linalg.solve(A,b))
print(np.linalg.inv(A)@b)
```

多项式

```python
p1 = np.poly1d([3, 2, 1])
p2 = np.poly1d([1, 2, 3])
print(p1)
print(p2)
```

```python
print(p1.coefficients)
print(p2.coeffs)
```

```python
print(p1 + p2)
print(p1 * p2)
```

```python
print(p1.deriv())
print(p1.integ())
```

```python
p3 = np.poly1d([1, 3, 2])
print(p3.roots)
```

```python
from numpy.polynomial import Polynomial

p3 = Polynomial((2, 3, 1))
print(p3)           # 输出多项式
print(p3(3))        # 令x=3，计算多项式的值
print(p3.roots())   # 计算多项式的根
print(p3.degree())  # 获得多项式的次数
print(p3.deriv())   # 求导
print(p3.integ())   # 求不定积分
```
