2.使用arange函数，指定取值范围和跨度创建备份对象
np.arange(start, stop, step, dtype=None)
np.arange(10),默认0-10步长为1

```python
array3=np.arange(0,20,2)
array3
```

3.使用linspace函数，用指定范围和元素个数创建备份对象，生成等差数列,公差就是（大-小）/n-1
np.linspace(start, stop, num=50, endpoint=True, dtype=None)

```python
array4=np.linspace(1,10,10,dtype='i8')
array4
```

4.使用logspace函数，生成等比数列
np.logspace(start, stop, num=50, endpoint=True, base=10.0)

```python
array5=np.logspace(0,10,11,base=2)
array5
```

5.通过fromstring函数从字符串导出创建备份对象

```python
array6=np.fromstring('1,2,3,4,5',sep=',',dtype='i8')
array6
```

6.通过fromiter函数从生成器（迭代器）中获取数据创建备份对象

```python
def fib(n):
    a,b=0,1
    for _ in range(n):
        a,b=b,a+b
        yield a
gen=fib(20)
array7=np.fromiter(gen,dtype='i8')
array7
```

- array.size:元素个数
- array.shape:数组形状
- array.ndim:数组维度
- array.dtype:元素数据类型
- array.itemsize:每个元素的大小，字节
- array.nbytes：数组总内存
- array.T:转置
- array.strides:步长（从一行，一列跳到下一行下一列所需要的字节）
- array.flags:它会显示数组在内存中是不是连续、是不是可写等信息。


```python
from IPython.display import Image, display

display(Image(filename="dtype.jpg"))
```

```python
print("元素个数:     ", array16.size)
# 输出：5024352
# 原因：1052 × 1592 × 3 = 5024352，图片每个像素有RGB三个数字，全部数字加起来的总个数

print("数组形状:     ", array16.shape)
# 输出：(1052, 1592, 3)
# 原因：图片高1052像素、宽1592像素、每个像素3个通道(RGB)

print("数组维度:     ", array16.ndim)
# 输出：3
# 原因：shape有3个数，对应3个维度（高度/宽度/通道）

print("元素数据类型: ", array16.dtype)
# 输出：uint8
# 原因：RGB值范围是0~255，刚好用uint8表示（无符号8位整数），不需要负数也不会超过255

print("每个元素大小: ", array16.itemsize, "字节")
# 输出：1 字节
# 原因：dtype是uint8，8位 = 1字节，所以每个数字占1字节

print("数组总内存:   ", array16.nbytes, "字节")
# 输出：5024352 字节
# 原因：nbytes = size × itemsize = 5024352 × 1 = 5024352字节 ≈ 4.8MB
```

随机生成图片

```python


# 随机生成一个 400x600 的彩色图片数组
random_img = np.random.randint(0, 256, (400, 600, 3), dtype=np.uint8)   


print(random_img.shape)   # (400, 600, 3)
print(random_img.dtype)   # uint8

# 显示出来
plt.imshow(random_img)
plt.axis('off')           # 不显示坐标轴
plt.show()
```
