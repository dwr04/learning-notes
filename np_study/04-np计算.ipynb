广播机制

```python
array5 = np.array([[0, 0, 0], [1, 1, 1], [2, 2, 2], [3, 3, 3]])
array6 = np.array([1, 2, 3])
"""
广播机制：
当两个数组 shape 不完全相同时，NumPy 会尝试通过“补 1”和“扩展维度为 1 的轴”，让它们形状一致。

规则：
1. 先把两个 shape 从右往左对齐。
2. 如果某个数组维度更少，就在左边补 1。
3. 从右往左逐维比较：
   - 两个维度相等，可以；
   - 或者其中一个维度为 1，可以；
   - 否则不能广播。
4. 所有维度都通过后，结果 shape 取每一维中较大的那个。

"""
print(array5 + array6)
```

```python
arr=np.arange(1,13).reshape(3,4)
np.extract(arr%2==0,arr)#前面筛选后面数组
```

```python
#np.select(condlist, choicelist, default=0)注意：多个条件同时满足时，选第一个
np.select([arr<= 3, arr>= 7], [arr * 10, arr ** 2])
```

```python
arr = np.array([1, 5, 8, 12, 20])

conditions = [
    arr < 5,
    (arr >= 5) & (arr < 10),
    arr >= 10
]

choices = [
    "小",
    "中",
    "大"
]

result = np.select(conditions, choices, default="未知")

print(result)
```
