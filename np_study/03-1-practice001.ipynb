```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("numpy_pra.jpeg")
arr = np.array(img)

print(type(img))
print(arr)
print(arr.shape)#(810,1080,3)
print(arr.dtype)

plt.imshow(arr)
plt.axis("off")
plt.show()
```

```python
plt.imshow(arr[::-1])
#0轴反向索引反转图片
```

```python
plt.imshow(arr)
#镜像图片
plt.imshow(arr[:,::-1])
```

```python
#抠图
plt.imshow(arr[305:505,440:640])
```

```python
#降采样
plt.imshow(arr[305:505:10,440:640:10])
```

```python
plt.imshow(arr[:,:,0],cmap='gray')
#白代表强黑代表弱
#只保留红色通道
```
