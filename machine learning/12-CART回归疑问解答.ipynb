
### 1. 寻找最佳切分点： SSE 等价于加权 MSE

**理论目标**：最小化左右子节点的加权均方误差（加权 MSE）。


$$Target = \frac{N_{left}}{N} MSE_{left} + \frac{N_{right}}{N} MSE_{right}$$

**代数化简**：代入定义 $MSE = \frac{SSE}{n}$：


$$Target = \frac{N_{left}}{N} \left( \frac{SSE_{left}}{N_{left}} \right) + \frac{N_{right}}{N} \left( \frac{SSE_{right}}{N_{right}} \right)$$

分子分母的样本数（$N_{left}$ 和 $N_{right}$）直接抵消：


$$Target = \frac{SSE_{left}}{N} + \frac{SSE_{right}}{N} = \frac{SSE_{left} + SSE_{right}}{N}$$

**结论**：因为在当前节点的循环比较中，总样本数 $N$ 是**固定常数**，所以：


$$\min(Target) \iff \min(SSE_{left} + SSE_{right})$$


即代码中直接找最小的 `sse_children`，完美等价，且省去了所有除法运算。

### 2. 特征重要度：为什么可以直接累加 SSE 减少量？

**理论公式**：单次切分的重要性增量，需考虑样本权重。


$$\Delta Importance = \frac{N_t}{N_{total}} \times \left( MSE_{parent} - \frac{N_{left}}{N_t} MSE_{left} - \frac{N_{right}}{N_t} MSE_{right} \right)$$

**代数化简**：把 $N_t$ 乘进括号内：


$$\Delta Importance = \frac{1}{N_{total}} \times \left( N_t \cdot MSE_{parent} - N_{left} \cdot MSE_{left} - N_{right} \cdot MSE_{right} \right)$$

将所有的 $N \cdot MSE$ 还原为 $SSE$：


$$\Delta Importance = \frac{1}{N_{total}} \times \left( SSE_{parent} - SSE_{left} - SSE_{right} \right) = \frac{parent\_sse - sse\_children}{N_{total}}$$

**结论**：理论真实增量 = `代码累加量` / $N_{total}$。由于全局总样本数 $N_{total}$ 是常数，代码中**仅累加分子**，待整棵树建完后再做一次全局归一化（算百分比），数学上完全等价。
